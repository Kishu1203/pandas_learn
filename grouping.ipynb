```python
import pandas as pd 

df = pd.read_csv('./data/Flavors.csv')
df
```

```python
g1 = df.groupby('Base Flavor')  # This groups the data by 'Base Flavor' column
g1.size()  # This gives the size of each group
g1.mean(numeric_only=True)  # Shows the mean of numeric columns for each group
g1.describe()  # Provides descriptive statistics for each group
g1.sum(numeric_only=True)  # Sums up numeric columns for each group
g1.min(numeric_only=True)  # Minimum value in numeric columns for each group
g1.max(numeric_only=True)  # Maximum value in numeric columns for each group
g1.count()


```

```python
g1.agg({'Flavor Rating': ['mean','count','max','min','sum'], 'Texture Rating': 'count' })  # aggregates multiple statistics for specified columns
```

```python
g2 = df.groupby(['Base Flavor', 'Flavor'])  # This groups the data by both 'Base Flavor' and 'Flavor' columns
g2.mean(numeric_only = True) # Shows the mean of numeric columns for each group
```
