```python
import pandas as pd
import numpy as np

df= pd.read_csv('./data/fighter_details.csv')
df.head(10)
```

```python
df.isnull()  # it gives the number of null values in each column
df.isna() # it gives the number of null values in each column

df.isnull().sum()  # it gives the total number of null values in each column

df.isna().sum().sum() # it gives the total number of null values in the entire dataframe
```

```python

# opposite of isnull() and isna()
df.notna() # it gives the number of non-null values in each column
df.notna().sum() # it gives the total number of non-null values in each column
df.notna().sum().sum() # it gives the total number of non-null values in the entire dataframe
```
