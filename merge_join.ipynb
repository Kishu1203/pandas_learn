Merge and Join and Concatenate

```python
import pandas as pd

df1= pd.read_csv('./data/LOTR.csv')
df1
```

```python
df2= pd.read_csv('./data/LOTR 2.csv')
df2
```

```python
df1.merge(df2)
```

```python
df1.merge(df2, how='inner', on='FellowshipID')  # Inner join on 'Name' column
df1.merge(df2, how='outer', on='FellowshipID')  # Outer join on 'Name' column
df1.merge(df2, how='left', on='FellowshipID') # Left join on 'Name' column
df1.merge(df2, how='right', on='FellowshipID') # Right join on 'Name' column

```

```python
df1.join(df2, how='left', on='FellowshipID', lsuffix = '_lkey', rsuffix = '_rkey')  # Inner join on 'Name' column
```

```python
# joining dataframes with index becase both DataFrames have 'FellowshipID' as an index we can use colname 'FellowshipID' to join them
df3 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), how='left', lsuffix = '_lkey', rsuffix = '_rkey')  # It joins on the index of both DataFrames

df3
```

```python
pd.concat([df1, df2])  # Concatenates the two DataFrames 
pd.concat([df1, df2], axis=1)  # Concatenates the two DataFrames along columns
pd.concat([df1, df2], axis=0, ignore_index=True)  # Concatenates the two DataFrames along rows and resets the index

pd.concat([df1, df2], axis=0, ignore_index=True, join = 'outer')  # Concatenates the two DataFrames along rows, resets the index, and performs an outer join
```
